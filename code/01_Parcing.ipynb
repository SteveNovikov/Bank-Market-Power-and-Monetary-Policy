```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
from progress.bar import IncrementalBar
from tqdm.auto import tqdm
```

```python
def new_row(regnum, dt_year, dt_month):
    temp_dict = {}
    
    #balance sheet
    url_f802 = "https://cbr.ru/banking_sector/credit/coinfo/f802/?regnum="+str(regnum)+"&dt="+str(dt_year)+dt_month
    r = requests.get(url_f802)
    if r.status_code == 200:
        soup = bs(r.text, "html.parser")
        table = soup.find('table', class_='data spaced')
        trs = table.find_all('tr')
        #print(trs)
        
        if table.find('td', class_='bold') is None:
            return {}
        
        temp_dict['regnum'] = regnum
        temp_dict['dt_year'] = dt_year
        temp_dict['dt_month'] = dt_month
        temp_dict[soup.find('div', class_='coinfo_item_title col-md-9').text] = soup.find('div', class_='coinfo_item_text col-md-13 offset-md-1').text

        for tr in trs:
            tds = tr.find_all('td')
            if len(tds) == 4:
                #print(tds)
                if tds[3].text.replace(" ", "") == '':
                    num = None
                else:
                    num = int(tds[3].text.replace(" ", ""))
                #print(tds[0].text, tds[1].text, num)
                temp_dict[tds[0].text +  ' ' + tds[1].text] = num
            elif len(tds) > 4:
                #print(tds)
                if tds[len(tds)-1].text.replace(" ", "") == '':
                    num = None
                else:
                    print(tds[len(tds)-1].text.replace(" ", ""))
                    num = int(tds[len(tds)-1].text.replace(" ", ""))
                #print(tds[0].text, tds[1].text, num)
                temp_dict[tds[0].text +  ' ' + tds[1].text] = num

    time.sleep(0.1) # sleep for 0.1 seconds
    #income statement
    url_f803 = "https://cbr.ru/banking_sector/credit/coinfo/f803/?regnum="+str(regnum)+"&dt="+str(dt_year)+dt_month
    r = requests.get(url_f803)
    if r.status_code == 200:
        soup = bs(r.text, "html.parser")
        table = soup.find('table', class_='data spaced')
        trs = table.find_all('tr')
        #print(trs)
        
        if table.find('td', class_='bold') is None:
            return {}
        
        temp_dict['regnum'] = regnum
        temp_dict['dt_year'] = dt_year
        temp_dict['dt_month'] = dt_month
        temp_dict[soup.find('div', class_='coinfo_item_title col-md-9').text] = soup.find('div', class_='coinfo_item_text col-md-13 offset-md-1').text

        for tr in trs:
            tds = tr.find_all('td')
            if len(tds) == 4:
                #print(tds)
                if tds[3].text.replace(" ", "") == '':
                    num = None
                else:
                    num = int(tds[3].text.replace(" ", ""))
                #print(tds[0].text, tds[1].text, num)
                temp_dict[tds[0].text +  ' ' + tds[1].text] = num
            elif len(tds) > 4:
                #print(tds)
                if tds[len(tds)-1].text.replace(" ", "") == '':
                    num = None
                else:
                    print(tds[len(tds)-1].text.replace(" ", ""))
                    num = int(tds[len(tds)-1].text.replace(" ", ""))
                #print(tds[0].text, tds[1].text, num)
                temp_dict[tds[0].text +  ' ' + tds[1].text] = num
    return temp_dict
```

```python
dt_year = 2023
dt_month = '09'

df = pd.DataFrame()
#bar = IncrementalBar('Countdown', max = len(pd.read_excel('BankList.xlsx')['cregnum']))
regnums = pd.read_excel('BankList.xlsx')['cregnum']
i = 0
for regnum in tqdm(regnums):
    row = new_row2(regnum, dt_year, dt_month)
    if row != {}:
        #df = df.append(row, ignore_index=True)
        df = pd.concat([df, pd.DataFrame(row, index=[i])], ignore_index=True)
    time.sleep(0.1)
    #bar.next()
    i += 1
#bar.finish()

df.to_excel("bankdata_"+str(dt_year)+"_"+dt_month+".xlsx")
```
